# Mecanum trajectory visualizer

Interactive widget for the KUKA youBot Mecanum digital-twin simulations.

**Data:** read-only Arrow files in `data/Simulation_Data_MecanumSlipSpin_LugreAdamov/`.
**Dropdowns:**
- `profile` — trajectory/excitation name (e.g. `octagon`, `coupled_vomega`).
- `mu` — friction coefficient.
- `chi` — LuGre adhesion coefficient.
- `combo` — sweep combo id (populated per profile).

**Visualization:**
- Dotted black line = reconstructed world-frame reference path.
- Blue oriented rectangle = KUKA youBot chassis (long axis = heading).
- Red pointer = heading (body +x).
- Solid blue trace = actual path up to the current time.

Use the slider or the Play button to scrub/animate time.  The Export GIF button writes a playback GIF under `trajectory_widget/exports/`.
